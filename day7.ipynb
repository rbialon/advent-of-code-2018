##### Part 1

```python
from collections import defaultdict

prerequisites = defaultdict(set)
solution = ""

for line in puzzle:
    line = line.split()
    prereq = line[1]
    step = line[7]
    
    prerequisites[step].add(prereq)
    prerequisites[prereq] = prerequisites[prereq]


while any(prerequisites):
    current_step = sorted(list(filter(lambda x: not prerequisites[x], prerequisites)))[0]
    solution += current_step
    
    for k in prerequisites:
        prerequisites[k].discard(current_step)
    
    prerequisites.pop(current_step)

print("Found solution {}".format(solution))
```

##### Part 2

```python
workers = dict()
for i in range(5):
    workers[i] = {"step": "", "counter": 0}

prerequisites = defaultdict(set)
timer = -1 # Starting at 0

for line in puzzle:
    line = line.split()
    prereq = line[1]
    step = line[7]
    
    prerequisites[step].add(prereq)
    prerequisites[prereq] = prerequisites[prereq]


while any(prerequisites) or any([workers[x]["counter"] >= 0 for x in workers]):
    timer += 1
    
    for w in workers:
        workers[w]["counter"] -= 1
    
    for w in filter(lambda x: workers[x]["counter"] < 0, workers):
        worker_step = workers[w]["step"]
        
        if worker_step:
            for k in prerequisites:
                prerequisites[k].discard(worker_step)
            
            workers[w]["step"] = ""

    for w in filter(lambda x: not workers[x]["step"], workers):
        current_steps = sorted(list(filter(lambda x: not prerequisites[x], prerequisites)))
        if current_steps:
            current_step = current_steps[0]
            workers[w]["step"] = current_step
            workers[w]["counter"] = 60 + ord(current_step) - 65
            prerequisites.pop(current_step)
                
    print("{}\t{}\t{}\t{}\t{}\t{}".format(timer, *list(map(lambda x: workers[x]["step"], workers))))

print("Found solution in {}s".format(timer))
```
