## Day 12

```python
puzzle = "##.#...#.#.#....###.#.#....##.#...##.##.###..#.##.###..####.#..##..#.##..#.......####.#.#..#....##.#"

rules_description = """#.#.# => #
..#.# => .
.#.## => #
.##.. => .
##... => #
##..# => #
#.##. => #
.#..# => #
.#### => .
....# => .
#.... => .
#.### => .
###.# => #
.#.#. => .
#...# => .
.#... => #
##.#. => #
#..## => #
..##. => .
####. => #
.###. => .
##### => .
#.#.. => .
...#. => .
..#.. => .
###.. => #
#..#. => .
.##.# => .
..... => .
##.## => #
..### => #
...## => #"""
```

##### Part 1

```python
pots = dict()
for i, pot in enumerate(puzzle):
    pots[i] = pot
    
rules = dict()
for rule in rules_description.splitlines():
    key, value = rule.split(" => ")
    rules[key] = value

def take_five(index, pots):
    five = ""
    for i in range(index - 2, index + 3):
        five += pots.get(i, ".")
        
    return five


for generation in range(20):
    changes = []
    
    for index in range(min(pots) - 2, max(pots) + 3):
        five = take_five(index, pots)
        if five in rules and five[2] != rules[five]:
            changes.append((index, rules[five]))
            
    for index, change in changes:
        pots[index] = change
        
        
solution = sum([x for x in pots if pots[x] == "#"])
print("Found solution {}".format(solution))
```

##### Part 2

```python
pots = dict()
for i, pot in enumerate(puzzle):
    pots[i] = pot
    
rules = dict()
for rule in rules_description.splitlines():
    key, value = rule.split(" => ")
    rules[key] = value

def take_five(index, pots):
    five = ""
    for i in range(index - 2, index + 3):
        five += pots.get(i, ".")
        
    return five

def pots_as_string(pots):
    return "".join(pots[x] for x in sorted(pots))

prev_sum = 0

for generation in range(150):#50000000000):
    changes = []
    
    for index in range(min(pots) - 2, max(pots) + 3):
        five = take_five(index, pots)
        if five in rules and five[2] != rules[five]:
            changes.append((index, rules[five]))
            
    for index, change in changes:
        pots[index] = change
        
    print(pots_as_string(pots))
    current_sum = sum([x for x in pots if pots[x] == "#"])
    print("Difference: {} current sum is {} for generation {}".format(current_sum - prev_sum, current_sum, generation))
    prev_sum = current_sum
```
