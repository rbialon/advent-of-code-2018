##### Part 1

```python
import numpy as np

tracks = np.array(list("".join(puzzle)), dtype=str)
tracks = tracks.view("U1").reshape((-1, len(puzzle[0])))

original_tracks = np.copy(tracks)
for x, y in np.argwhere(np.isin(original_tracks, list("^v"))):
    original_tracks[x, y] = "|"
for x, y in np.argwhere(np.isin(original_tracks, list("<>"))):
    original_tracks[x, y] = "-"


corners = list("/\\")
path = list("-|")
intersection = list("+")
carts = list("^v<>")

directions = {
    "^": (-1, 0),
    "v": (1, 0),
    "<": (0, -1),
    ">": (0, 1)
}

corner_directions = {
    ("^", "\\"): "<",
    ("^", "/"): ">",
    ("v", "\\"): ">",
    ("v", "/"): "<",
    ("<", "\\"): "^",
    ("<", "/"): "v",
    (">", "\\"): "v",
    (">", "/"): "^"
}

intersection_directions = {
    0: "\\",
    1: "_",
    2: "/"
}

intersection_invert = {
    "<": ">",
    ">": "<"
}


carts_turn_counter = dict()
for x, y in np.argwhere(np.isin(tracks, carts)):
    carts_turn_counter[(x, y)] = 0

ticks = 0
while not np.any(tracks == "X"):

    for x, y in sorted([(x, y) for x, y in np.argwhere(np.isin(tracks, carts))]):
        turn_counter = carts_turn_counter.pop((x, y))
        original_turn_counter = turn_counter
        
        original_track = original_tracks[x, y]
        cart = tracks[x, y]
        
        if original_track in path:
            new_cart_symbol = cart
            
        elif original_track in corners:
            new_cart_symbol = corner_directions[(cart, original_track)]
            
        elif original_track in intersection:
            intersection_direction = intersection_directions[turn_counter % 3]
            turn_counter += 1
            
            intersection_cart = intersection_invert.get(cart, cart)
            new_cart_symbol = corner_directions.get((intersection_cart, intersection_direction), cart)
            
        else:
            print("error: {}".format(original_track))
            break
            
            
        direction = directions[new_cart_symbol]
        new_cart = (x + direction[0], y + direction[1])
        
        if tracks[new_cart] in carts:
            tracks[new_cart] = "X"
            tracks[x, y] = original_track
            print("First Crash at {}".format(new_cart[::-1]))
            break
        
        tracks[new_cart] = new_cart_symbol
        tracks[x, y] = original_track
        carts_turn_counter[new_cart] = turn_counter
    
    ticks += 1
    #print("\n".join(map(lambda x: "".join(x), tracks)))
```

```python
import numpy as np

tracks = np.array(list("".join(puzzle)), dtype=str)
tracks = tracks.view("U1").reshape((-1, len(puzzle[0])))

original_tracks = np.copy(tracks)
for x, y in np.argwhere(np.isin(original_tracks, list("^v"))):
    original_tracks[x, y] = "|"
for x, y in np.argwhere(np.isin(original_tracks, list("<>"))):
    original_tracks[x, y] = "-"


corners = list("/\\")
path = list("-|")
intersection = list("+")
carts = list("^v<>")

directions = {
    "^": (-1, 0),
    "v": (1, 0),
    "<": (0, -1),
    ">": (0, 1)
}

corner_directions = {
    ("^", "\\"): "<",
    ("^", "/"): ">",
    ("v", "\\"): ">",
    ("v", "/"): "<",
    ("<", "\\"): "^",
    ("<", "/"): "v",
    (">", "\\"): "v",
    (">", "/"): "^"
}

intersection_directions = {
    0: "\\",
    1: "_",
    2: "/"
}

intersection_invert = {
    "<": ">",
    ">": "<"
}


carts_turn_counter = dict()
for x, y in np.argwhere(np.isin(tracks, carts)):
    carts_turn_counter[(x, y)] = 0

ticks = 0
while np.count_nonzero(np.isin(tracks, carts)) > 1:

    current_carts = sorted([(x, y) for x, y in np.argwhere(np.isin(tracks, carts))])
    while current_carts:
        x, y = current_carts.pop(0)
        turn_counter = carts_turn_counter.pop((x, y))
        original_turn_counter = turn_counter
        
        original_track = original_tracks[x, y]
        cart_symbol = tracks[x, y]
        
        if original_track in path:
            new_cart_symbol = cart_symbol
            
        elif original_track in corners:
            new_cart_symbol = corner_directions[(cart_symbol, original_track)]
            
        elif original_track in intersection:
            intersection_direction = intersection_directions[turn_counter % 3]
            turn_counter += 1
            
            intersection_cart_symbol = intersection_invert.get(cart_symbol, cart_symbol)
            new_cart_symbol = corner_directions.get((intersection_cart_symbol, intersection_direction), cart_symbol)
            
        else:
            print("error: {}".format(original_track))
            break
            
            
        direction = directions[new_cart_symbol]
        new_cart = (x + direction[0], y + direction[1])
        
        if tracks[new_cart] in carts:
            new_cart_symbol = original_tracks[new_cart]
            carts_turn_counter.pop(new_cart)
            if new_cart in current_carts:
                current_carts.remove(new_cart)

        else:
            carts_turn_counter[new_cart] = turn_counter
            
        tracks[new_cart] = new_cart_symbol
        tracks[x, y] = original_track
    
    ticks += 1
    #print("\n".join(map(lambda x: "".join(x), tracks)))
    
print("Found last cart left at {}".format(np.argwhere(np.isin(tracks, carts))[0][::-1]))
```
